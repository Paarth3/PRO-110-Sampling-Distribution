```python
import pandas as pd
import plotly.figure_factory as ff
import random
import  statistics
```

```python
df = pd.read_csv("C:/Users/ADMIN/Desktop/WhiteHat_Jr/General/Data/medium_data.csv")
```

```python
population_mean = statistics.mean(df["reading_time"])
```

```python
def random_30():
    random_list = []
    for i in range(0,30):
        random_num = random.choice(df["reading_time"])
        random_list.append(random_num)
    return statistics.mean(random_list)
```

```python
def setup():
    mean_of_samples = []
    for i in range(0,100):
        random_mean = random_30()
        mean_of_samples.append(random_mean)
    return mean_of_samples
```

```python
def plot(mean_list):
    fig = ff.create_distplot([mean_list], ["Reading Time"], show_hist=False)
    fig.show()
```

```python
mean_list = setup()
figure = plot(mean_list)
```
